```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns 
import glob 
```

```python
files = glob.glob('*.csv')
```

```python
dfs = []
for file in files:
    df = pd.read_csv(file)
    dfs.append(df)
    
```

```python
pitcher = pd.concat(dfs, sort=False)
pd.options.display.max_columns = 100
pd.options.display.max_rows = 300
```
